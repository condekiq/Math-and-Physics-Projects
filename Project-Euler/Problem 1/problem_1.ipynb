```python
# https://projecteuler.net/problem=1

multiple_of_3 = []
multiple_of_3_and_5 = [] 
multiple_of_5 = []

for i in range(1000):
	if i%3==0 and i%5==0:
		multiple_of_3_and_5.append(i)

	elif i%3==0:
		multiple_of_3.append(i)

	elif i%5==0:
		multiple_of_5.append(i)

answer = sum(multiple_of_3 + multiple_of_3_and_5 + multiple_of_5)
answer
```
